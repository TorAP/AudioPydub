## Initial setup
**f Relevant:** takes in all the correlated values and only returns those, where the absolute value (distance from 0) is higher than set (0.5)

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


def relevant(s):
    out = [abs(x) if abs(x) > 0.66 else 0 for x in s]
    return pd.Series(out)


df = pd.read_excel("Questionnaire.xlsx", sheet_name="Responses")
df.drop("#", inplace=True, axis=1)
```

**Original correlation ranging from purple (-1) to green (1)**

```python
corr = df.corr(method='spearman')
corr.style.background_gradient(cmap='PiYG', axis=None).set_precision(2)
```

**Correlation with relevance applied on absolute values, ranging from white (0) to green (1)**

```python
corr.apply(relevant).style.background_gradient(cmap='Greens', axis=None).set_precision(2)
```
